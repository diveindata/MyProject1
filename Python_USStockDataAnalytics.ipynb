```
import yahoo_fin.stock_info as yf
import pandas as pd
import matplotlib.pyplot as plt
plt.style.use('fivethirtyeight')
plt.rcParams['figure.figsize'] = [24, 8]

symbol = "MSFT"

# Get historical data
hist_data = yf.get_data(symbol)

# Create subplots for the charts
fig, axes = plt.subplots(nrows=2, ncols=1, figsize=(24, 16))

# Plotting closing prices
hist_data['close'].plot(ax=axes[0], color='blue')
axes[0].set_title('Historical Closing Prices')

# Plotting volume
hist_data['volume'].plot(ax=axes[1], color='green')
axes[1].set_title('Historical Volume')

# Show the plots
plt.tight_layout()
plt.show()

```
